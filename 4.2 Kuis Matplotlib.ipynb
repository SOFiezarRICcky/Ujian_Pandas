```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Soal 1
Dari data x dan y buatlah sebuah figure dengan ketentuan :
1. Judul figure adalah Data Penjualan PT. Sejahtera
2. X label adalah Pembelian
3. Y label adalah penjualan

```python
plt.plot(x,y,"b")
plt.xlabel("Pembelian")
plt.ylabel("Penjualan")
plt.title("Data Penjualan PT. Sejahtera")
plt.show()
```

## Soal 2

Buatlah sebuah figure dengan ketentuan :
1. menggunakan 2 axes
2. ax1 adalah [0,0,1,1] dan ax2 adalah [0.2,0.5,.2,.2]

```python
fig, main_ax = plt.subplots()
main_ax.set_xlim(0,1)
main_ax.set_ylim(0,1)

right_insert_ax = fig.add_axes([0.2,0.5,.2,.2])




plt.show()
```

sekarang isi data dari masing - masing axes tersebut dengan x dan y

```python
fig, main_ax = plt.subplots()
main_ax.plot(x,y)

plt.xlabel("Sumbu X")
plt.ylabel("Sumbu Y")
plt.title("Soal 1")

right_insert_ax = fig.add_axes([0.25,0.5,.2,.2])
right_insert_ax.plot(x,y)
right_insert_ax.set_xlabel("x")
right_insert_ax.set_ylabel("y")

plt.show()
```

## Soal 3

Buatlah sebuah figure dengan ketentuan :
1. Menggunakan sub plot 1 baris 2 kolom

```python
plt.subplot(1,2,1)

plt.subplot(1,2,2)
plt.show()
```

2. Sekarang isi data plot pertama dengan data x dan y dan buat sesuai dengan tampilan di bawah
3. Sekarang isi data plot pertama dengan data x dan z dan buat sesuai dengan tampilan di bawah

```python
import numpy as np
plt.subplot(1,2,1)

plt.plot(x,y,'b*-')

plt.subplot(1,2,2)

plt.plot(x,z,"b*-")
```

Buat tampilannya menjadi seperti ini

```python

fig, axs = plt.subplots(1,2, constrained_layout=True, figsize=(14,2))

axs[0].set_xlabel("x")
axs[0].set_ylabel("y")
axs[0].plot(x,y,"b*-")

axs[1].set_xlabel("x")
axs[1].set_ylabel("y")
axs[1].plot(x,z,"b*-")
plt.show()
```
